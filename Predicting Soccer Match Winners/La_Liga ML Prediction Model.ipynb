```python
#converting existing column to date time and overriding the existing column with the date time.
matches["Date"] = pd.to_datetime(matches["Date"])
```

```python
#create few basic predictors, foundation for our model. 
#venue code converting home and away column into a numberic column (important predictor: home field advantage)

matches["Venue_Code"] = matches["Venue"].astype("category").cat.codes #convert string into category then category into numbers
matches["Opp_Code"] = matches["Opponent"].astype("category").cat.codes
matches["Hour"] = matches["Time"].str.replace(":.+", "", regex= True).astype("int")
matches["Day_Code"] = matches["Date"].dt.dayofweek
```

```python
#Turning our W's into 1 and L and D into 0's
matches["Target"] = (matches["Result"] == "W").astype("int") 
```

```python
#Creating the ML Learning Model using SKLearn and RandomForest Classifier
```

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
#call randomforest, pass in paramterers, 
#n_estimators number of desicion trees we want to train
#higher the number is longer it takes to run but more accurate it is. 
#min_sample_split number of samples we want to have in leaf of desicion tree before we split node.
# higher it is, less likely to overfit, but less accuracy data will be. 
#random state if we run forerst mulitple times we get the same results as long as data is the same. 
```

```python
rf = RandomForestClassifier(n_estimators = 50, min_samples_split= 10, random_state = 1)
```

```python
#Train data from the past to predict out comes of the future, can't predict future for data of past!!
```

```python
train_model = matches[matches["Date"]<'2023-01-01'] #check this
test_set = matches[matches["Date"]>'2022-09-01'] #check this 
```

```python
predictors = ["Venue_Code", "Opp_Code", "Day_Code", "Hour"]
```

```python
#fit our randomforest model using rf.fit: trains our forest model with predictors 

rf.fit(train_model[predictors], train_model["Target"])
```

```python
preds = rf.predict(test_set[predictors])
```

```python
#test metrics and accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(test_set["Target"],preds)
```

```python
#make function called rolling_averages: take a group of old columns to assign rollinga verages for and new columns we want to assign rolling averages to.  

def rolling_averages(group,cols,new_cols):
    group = group.sort_values("Date")
    rolling_stats = group[cols].rolling(3, closed ='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset = new_cols)
    return group
```

```python
cols = ["GF","GA","Sh","SoT","Dist","FK","PK","PKatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group,cols,new_cols)
```

```python
#took original matches data frame and applied a function to each of those team data frames to compute the rolling averages

matches_rolling = matches.groupby("Team").apply(lambda x:rolling_averages(x,cols,new_cols))
```

```python
#new predictors to make a new sense of predictions to retrain our model
```

```python
def make_predictions(data, predictors):
    train = data[data["Date"] < '2023-01-01']
    test = data[data["Date"] > '2022-09-01']
    rf.fit(train[predictors], train["Target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["Target"], predicted=preds), index=test.index)
    precision = precision_score(test["Target"], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```
